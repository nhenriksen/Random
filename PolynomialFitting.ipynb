```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
### Random test case with no correlation

### Draw 25 random points from a normal distribution
N = 25
x = np.arange(0,N,1)
y = np.random.normal(-200,1,N)

### Get coefficients for a 9th degree polynomial fit
fitcoeffs = np.polyfit(x,y,9)
### Create polynomial function with the coefficients
fitp = np.poly1d(fitcoeffs)
### Generate polynomial fit line
xf = np.linspace(0,N-1,N*40)
yf = fitp(xf)

### Plot results
plt.plot(x,y,'bo')
plt.plot(xf,yf,'r-')
```

```python
### Test case with some correlation between i,i-1,and i-2

### Draw 25 random points, but average with i-1 and i-2
N=25
prev2 = -200.0  # Initialize at -200
prev1 = -200.0  # Initialize at -200
x = np.arange(0,N,1)
y = np.zeros([N],np.float64)
for i in range(N):
    y[i] = (np.random.normal(prev2,1) + np.random.normal(prev1,1) + np.random.normal(-200,1))/3
    prev1 = y[i]
    if i > 0:
        prev2 = y[i-1]

### Get coefficients for a 9th degree polynomial fit
fitcoeffs = np.polyfit(x,y,9)
### Create polynomial function with the coefficients
fitp = np.poly1d(fitcoeffs)
### Generate polynomial fit line
xf = np.linspace(0,N-1,N*40)
yf = fitp(xf)

### Plot results
plt.plot(x,y,'bo')
plt.plot(xf,yf,'r-')
```
